```python
# Using .loc (more explicit, handles both rows and cols at once)
a=df.loc[df['origin'] == 'Asia', 'horsepower'].max()

# Using .query (SQL-ish syntax, nice for complex filters)
b=df.query("origin == 'Asia'")['fuel_efficiency_mpg'].max()
print(a, b)
```

```python
# Median value of horsepower
naValues = df['horsepower'].isna()
print(f"Number of missing values in horsepower column: {naValues.sum()}")
median_hp = df['horsepower'].median() 
print(f"Median horsepower before filling missing values: {median_hp}")
mode_hp = df['horsepower'].mode()[0]
print(f"Mode horsepower before filling missing values: {mode_hp}")
# Fill missing values with mode and calculate median again
print(f"median horsepower after filling missing values: {df['horsepower'].fillna(mode_hp).median()}")
```

```python
cars_from_asia = df[df['origin'] == 'Asia']
cars_from_asia.shape[0]
cars_from_asia.isna().sum()
```

```python
cars_selection=cars_from_asia[['vehicle_weight', 'model_year']].iloc[:7,:]
cars_selection
```

```python
X = cars_selection.to_numpy()
X.shape    # (7, 2)
```

```python
XTX = np.matmul(X.T, X) 
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = XTX_inv @ X.T @ y
w.sum()
```
